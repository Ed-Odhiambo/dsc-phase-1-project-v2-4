Some case examples are the columns that are useless to the problem statements. They can be done away with in general. We can drop the column and check the shape to confirm the new data frame that should have less columns after.

```python
# Dropping our unnecessary columns in TMDB and confirming through its shape
TMDB_data.drop(['Unnamed: 0','original_language', 'original_title'],axis=1, inplace=True)
print("Shape of IMDB_data: ", TMDB_data.shape)
print("Columns now in IMDB_data: ", list(TMDB_data.columns))
```

```python
"""
In order to check for duplicated though-out the whole data, we are going to use a method that can draw out a boolean.
This is a True or False response where any mention of True for unique values indicates a duplicate is present.
"""

TMDB_data.duplicated()
set(TMDB_data.duplicated())
```

```python
# Dropping the duplicates from the data frame
TMDB_data.drop_duplicates(inplace=True)
TMDB_data.shape
```

```python
# Dropping rows in IMDB with less than 2000 vote counts or less that 6.0 average rating
for x in TMDB_data.index:
    if TMDB_data.loc[x, "vote_count"] < 2000 or TMDB_data.loc[x, "vote_average"] < 6.0:
        TMDB_data.drop(x, inplace=True)
        
TMDB_data
```

```python
# Getting the top 100 vote averages
TMDB_TopAvgVoted = TMDB_data.sort_values(by='vote_average', ascending=False).head(100)
TMDB_TopAvgVoted
```

```python
# Checking for duplicates in The Numbers data
TheNumbers_data.duplicated()
set(TheNumbers_data.duplicated())
```

```python
# Converting The Numbers data to integers for descriptive analysis
# Converting for the Production Budget column
TheNumbers_data['production_budget'] = TheNumbers_data['production_budget'].str.replace('[$,]', '', regex=True)
TheNumbers_data = TheNumbers_data.astype({"production_budget": int})

# Converting for the Domestic gross column
TheNumbers_data["domestic_gross"] = TheNumbers_data["domestic_gross"].str.replace('[$,]','', regex=True)
TheNumbers_data["domestic_gross"] = TheNumbers_data["domestic_gross"].astype(str).astype(int)

# Converting for the Worldwide Gross column
TheNumbers_data["worldwide_gross"] = TheNumbers_data["worldwide_gross"].str.replace('[$,]','', regex=True)
TheNumbers_data["worldwide_gross"] = TheNumbers_data["worldwide_gross"].astype(int)

# Printing out the data types
print(TheNumbers_data.dtypes)

# Calling the data frame
TheNumbers_data
```

## Data Modeling

What has been done is to remove the movies that have vote counts of less than 2000.
```python
# Genre vs revenue vs vote average: This combination could help answer question 3 
# by exploring which genres tend to be both profitable and high quality.
# A bubble chart or scatter plot could be used to visualize this relationship.

fig, ax = plt.subplots(figsize=(10,6))
sns.barplot(x=TMDB_TopAvgVoted['genre_ids'].head(20), y=TMDB_TopAvgVoted['vote_average'].head(20))
plt.title('Top Voted Genres')
plt.xlabel('Vote Averages')
plt.ylabel('Genres')
plt.yticks(rotation =20)
plt.show()
```

```python
# Release date vs vote average: This combination could help answer question 2 by exploring how the quality of movies has changed over time.
# A line chart or scatter plot could be used to visualize this relationship.
fig, ax = plt.subplots(figsize=(10,6))
plt.plot(TMDB_TopAvgVoted['release_date'].sort_values(ascending=False).head(20), TMDB_TopAvgVoted['vote_average'].head(20))
plt.title('Quality Changes Over Time')
plt.xlabel('Release Date')
plt.ylabel('Average Votes')
plt.xticks(rotation=20)
plt.show()
```

```python
# Popularity vs release date: This combination could be used to explore how popularity of movies varies over time.
#A line chart or scatter plot could be used to visualize this relationship.
fig, ax = plt.subplots(figsize=(10,6))
plt.plot(TMDB_TopAvgVoted['release_date'].sort_values(ascending=False).head(20), TMDB_TopAvgVoted['popularity'].sort_values(ascending=False).head(20))
plt.title('Quality Changes Over Time')
plt.xlabel('Release Date')
plt.ylabel('Popularity')
plt.xticks(rotation=20)
plt.show()
```

```python
# Genre vs popularity: This combination could be used to explore which genres tend to be more popular among moviegoers. 
# A bar chart or stacked bar chart could be used to visualize this relationship.
fig, ax = plt.subplots(figsize=(10,6))
plt.bar(TMDB_TopAvgVoted['genre_ids'].head(20), TMDB_TopAvgVoted['popularity'].sort_values(ascending=False).head(20))
plt.title('Popular Genres')
plt.xlabel('Genres')
plt.ylabel('Popularity')
plt.xticks(rotation=90)
plt.show()
```

```python
# Title vs popularity: This combination could be used to explore which movie titles tend to be more popular among moviegoers.
# A word cloud or bubble chart could be used to visualize this relationship
fig, ax = plt.subplots(figsize=(10,6))
plt.bar(TMDB_TopAvgVoted['title'].head(20), TMDB_TopAvgVoted['popularity'].sort_values(ascending=False).head(20))
plt.title('Popular Movies')
plt.xlabel('Titles')
plt.ylabel('Popularity')
plt.xticks(rotation=90)
plt.show()
```

```python

# Genre vs vote average: This combination could be used to explore how the quality of movies varies by genre.
# A box plot or violin plot could be used to visualize this relationship.
fig, ax = plt.subplots(figsize=(10,6))
sns.barplot(x=TMDB_TopAvgVoted['genre_ids'].head(20), y=TMDB_TopAvgVoted['vote_average'].sort_values(ascending=False).head(20))
plt.title('Quality Changes with Genre')
plt.xlabel('Genres')
plt.ylabel('Vote Averages')
plt.xticks(rotation=90)
plt.show()
```
