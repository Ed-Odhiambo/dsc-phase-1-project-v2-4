What has been done above is to use a method to get the general information of our dataframes. From here it is possible to tell what the data entails. From the information, it is seen that both sets have complete data with no missing values but that does not mean our data has accurate values. This means we would have to check for other measures. In the analysis it may not need some of these columns thus they can be done away with such as the unnamed, original_language and original_title columns in 'tmdb_df'. We can drop the column and check the shape to confirm. If the code has worked it should output one less column i.e from 10 columns to 7 columns.

```python
# Dropping our unnecessary columns in IMDB and confirming through its shape
IMDB_data.drop(['Unnamed: 0','original_language', 'original_title'],axis=1, inplace=True)
print("Shape of IMDB_data: ", IMDB_data.shape)
print("Columns now in IMDB_data: ", list(IMDB_data.columns))
```

```python
"""
In order to check for duplicated though-out the whole data, we are going to use a method that can draw out a boolean.
This is a True or False response where any mention of True for unique values indicates a duplicate is present.
"""

IMDB_data.duplicated()
set(IMDB_data.duplicated())
```

```python
# Dropping the duplicates from the data frame
IMDB_data.drop_duplicates(inplace=True)
IMDB_data.shape
```

```python
# Dropping rows in IMDB with less than 2000 vote counts and less that 6.0 average rating
for x in IMDB_data.index:
    if IMDB_data.loc[x, "vote_count"] < 2000 or IMDB_data.loc[x, "vote_average"] < 6.0:
        IMDB_data.drop(x, inplace=True)
        
IMDB_data
```

```python
IMDB_TopAvgVoted = IMDB_data.sort_values(by='vote_average', ascending=False).head(100)
IMDB_TopAvgVoted
```

```python
fig, axes = plt.subplots(figsize=(10,8))

sns.barplot(x =IMDB_TopAvgVoted['title'].head(10), y =IMDB_TopAvgVoted['vote_average'].head(10))
plt.title('Average Voted')
plt.xlabel('Movie')
plt.ylabel('Vote average')
plt.xticks(rotation = 90)

plt.show()
```

```python
IMDB_TopPopular = IMDB_data.sort_values(by='popularity', ascending=False).head(100)
IMDB_TopPopular
```

```python
fig, axes = plt.subplots(figsize=(10,8))

sns.barplot(x =IMDB_TopPopular['title'].head(10), y =IMDB_TopPopular['popularity'].head(10))
plt.title('Popularity')
plt.xlabel('Title')
plt.ylabel('Popular Charts')
plt.xticks(rotation = 90)

plt.show()
```

What has been done is to remove the movies that have vote counts of less than 2000. This is so that we can remain with data that has a good viewership as a good number of people watched and voted on the movie. This will help us analyse what has contributed to the number of votes if possible. Going ahead and limiting the movies to those with ratings of 6.0 and above is so that we can weed put non-performing movies and remain with those that could help give an idea of those that have the big companies behind them. This would help to see if we could use similar methods to get there.

```python
# Checking for duplicates in The Numbers data
TheNumbers_data.duplicated()
set(TheNumbers_data.duplicated())
```

```python
# Converting The Numbers data to integers for descriptive analysis
# Converting for the Production Budget column
TheNumbers_data['production_budget'] = TheNumbers_data['production_budget'].str.replace('[$,]', '', regex=True)
TheNumbers_data = TheNumbers_data.astype({"production_budget": int})

# Converting for the Domestic gross column
TheNumbers_data["domestic_gross"] = TheNumbers_data["domestic_gross"].str.replace('[$,]','', regex=True)
TheNumbers_data["domestic_gross"] = TheNumbers_data["domestic_gross"].astype(str).astype(int)

# Converting for the Worldwide Gross column
TheNumbers_data["worldwide_gross"] = TheNumbers_data["worldwide_gross"].str.replace('[$,]','', regex=True)
TheNumbers_data["worldwide_gross"] = TheNumbers_data["worldwide_gross"].astype(int)

# Printing out the data types
print(TheNumbers_data.dtypes)

# Calling the data frame
TheNumbers_data
```

```python
# Filter for movies with reasonable budgets with high returns
TheNumbers_profitable = TheNumbers_data.loc[TheNumbers_data['production_budget'] > 100000]
TheNumbers_profitable       
```

```python
TheNumbers_profitable = TheNumbers_profitable.loc[TheNumbers_profitable["worldwide_gross"] > 100000000]
TheNumbers_profitable 
```

```python
TheNum_Budgetable = TheNumbers_profitable.sort_values(by='production_budget').head(100)
TheNum_Budgetable
```

```python
fig, axes = plt.subplots(figsize=(10,8))

sns.barplot(x =TheNum_Budgetable['movie'].head(10), y =TheNum_Budgetable['production_budget'].head(10))
plt.title('Average Voted')
plt.xlabel('Movie')
plt.ylabel('Budget movies')
plt.xticks(rotation = 90)

plt.show()
```

```python
TheNum_TopGross = TheNumbers_profitable.sort_values(by='worldwide_gross', ascending=False).head(100)
TheNum_TopGross
```

```python
fig, axes = plt.subplots(figsize=(10,8))

sns.barplot(x =TheNum_TopGross['movie'].head(10), y =TheNum_TopGross['worldwide_gross'].head(10))
plt.title('Top Gross')
plt.xlabel('Movie')
plt.ylabel('Gross Levels')
plt.xticks(rotation = 90)

plt.show()
```
