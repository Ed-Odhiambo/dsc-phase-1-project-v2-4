```python
# Dropping our unnecessary columns in tmdb_df and confirming through its shape
tmdb_df.drop(['Unnamed: 0','original_language', 'original_title'],axis=1, inplace=True)
print("Shape of tmdb_df: ", tmdb_df.shape)
print("Columns now in tmdb_df: ", list(tmdb_df.columns))
```

```python
"""
In order to check for duplicated thoughout the whole data, we are going to use a method that can draw out a boolean.
This is a True or False response where any mention of True for unique values indicates a duplicate is present.
"""

tmdb_df.duplicated()
set(tmdb_df.duplicated())
```

```python
# Dropping the duplictates from the dataframe
tmdb_df.drop_duplicates(inplace=True)
tmdb_df.shape
```

```python
tn_df.duplicated()
set(tn_df.duplicated())
```

```python
# Dropping rows in tmdb_df with less than 2000 vote counts
for x in tmdb_df.index:
    if tmdb_df.loc[x, "vote_count"] < 2000:
        tmdb_df.drop(x, inplace=True)
        
tmdb_df
```

```python
# Dropping rows in tmdb_df with rating below 6.0
for x in tmdb_df.index:
    if tmdb_df.loc[x, "vote_average"] < 6.0:
        tmdb_df.drop(x, inplace=True)
        
tmdb_df
```

What has been done is to remove the movies that have vote counts of less than 2000. This is so that we can remain with data that has a good viewership as a good number of people watched and voted on the movie. This will help us analyse what has contributed to the number of votes if possible. Going ahead and limiting the movies to those with ratings of 6.0 and above is so that we can weed put non-performing movies and remain with those that could help give an idea of those that have the big companies behind them. This would help to see if we could use similar methods to get there.

```python
tn_df["production_budget"] = tn_df["production_budget"].str.replace(r'\D', '')
tn_df["production_budget"] = tn_df["production_budget"].astype(int)
print(tn_df.dtypes, "\n")
tn_df
```

```python
# Stripping characters from our tn_df and coverting the values to intergers
tn_df["domestic_gross"] = tn_df["domestic_gross"].str.replace(r'\D', '')
tn_df["domestic_gross"] = tn_df["domestic_gross"].astype(int)
print(tn_df.dtypes, "\n")
tn_df
```

```python
tn_df["worldwide_gross"] = tn_df["worldwide_gross"].str.replace(r'\D', '')
tn_df["worldwide_gross"] = tn_df["worldwide_gross"].astype(int)
print(tn_df.dtypes, "\n")
tn_df
```
